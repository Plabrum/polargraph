```python
def spiral(t):
    x = t/10*np.cos(t)
    y = t/10*np.sin(t)
    return (x,y)
```

```python
t  = np.linspace(0,400, 10000)
x, y = spiral(t)
```

```python
def polar_to_cartesian(angle, radius):
    return (radius*np.sin(angle), radius*np.cos(angle))
```

```python
def draw_polar(angle, radius, ax=None):
    '''
    Convert polar cordinate array into cartesian
    '''
    if not ax:
        fig,ax = plt.subplots(figsize=(4,4))
        ax.set_aspect('equal', adjustable='box')
    x,y = polar_to_cartesian(angle, radius)
    ax.plot(x, y, linewidth=0.1)
    return ax
```
