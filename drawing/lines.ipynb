```python
def draw_lines(no_lines, line_equation, colormap, save=False):
    fig, ax = plt.subplots(figsize=(5,5))
    ax.axis('off')
    lines = 100
    for i in range(no_lines):
        x = np.linspace(0,100,10000)
        y = line_equation(x,i)
        ax.plot(x, y, linewidth=0.4, color=colormap(i/no_lines))
    if save:
        fig.savefig("exports/export.png", dpi=600)
```

```python
def equation(x, i): 
    return 3*(i) + np.sin((30*x)/(i+1) )
    
draw_lines(100, equation, mpl.colormaps['winter'])
```
